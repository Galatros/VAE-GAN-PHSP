```python
import numpy as np
import tqdm as tqdm
import sklearn.datasets as skd
import math
from datetime import datetime
import os

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kde
from HDF5Generator import HDF5DatasetGenerator
```

```python
DATA_FILE = '../data/dataNormalized_10000000_6.npy'
KEYS = ['Ekine', 'X', 'Y', 'dX', 'dY', 'dZ']
real = np.load(DATA_FILE)
```

```python
outputHDFFilePathTrList = ['../data/DISP_0.5_ANGLE_0/HDF5/a1TrainGenerator.hdf5', '../data/DISP_0_ANGLE_0/HDF5/d3TrainGenerator.hdf5']
trainGen = HDF5DatasetGenerator(outputHDFFilePathTrList,1000000)
gen = trainGen.generator()
part = next(gen)
```

```python
fig,axs = plt.subplots(2,3)
keys = KEYS
for i, j in enumerate(keys):
    axs.flatten()[i].hist(part[:, i],100,alpha=0.5,density=True)
    axs.flatten()[i].set_title(j)
fig = plt.gcf()
fig.set_size_inches(27.5, 15.5)
plt.show()
```

```python
fig,axs = plt.subplots(2,3)
keys = KEYS
for i, j in enumerate(keys):
    axs.flatten()[i].hist(real[:, i],100,alpha=0.5,density=True)
    axs.flatten()[i].set_title(j)
fig = plt.gcf()
fig.set_size_inches(27.5, 15.5)
plt.show()
```
